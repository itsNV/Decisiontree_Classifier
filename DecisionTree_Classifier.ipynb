## Decision Tree Classifier

### Import the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.tree import DecisionTreeClassifier
```

### Import the Dataset

> we are importing the iris dataset from `sklearn.datasets` which has the training dataset and the teting dataset included in that

```python
from sklearn.datasets import load_iris
iris = load_iris()
print(iris)
```

```python
# training dataset
train = iris.data
test = iris.target

train.shape,test.shape
```

### Split the dataset

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test  = train_test_split(train,test,test_size=0.2)
```

### Post-Prunning : Create a tree than apply the prunning

#### Create a treemodel

```python
treemodel = DecisionTreeClassifier()
```

#### Fit the treemodel to create the tree

```python
from sklearn import tree
treemodel.fit(X_train,y_train)
```

#### See the tree

```python
from  sklearn import tree
tree.plot_tree(treemodel)
```

#### now do prunning  :  

> we have the 3 values at the output :

* set max_depth to 2 because we get 1st result at 1st level left node

* other two results we get at the 2nd level right ide 2 nodes

#### Updated treemodel

```python
treemodel_updated = DecisionTreeClassifier(max_depth=2)
treemodel_updated.fit(X_train,y_train)
```

#### Plot the tree again

```python
tree.plot_tree(treemodel_updated)
```

#### Now predict the test data

```python
y_pred = treemodel_updated.predict(X_test)
```

```python
y_pred
```

#### Find the accuracy and Report

```python
from sklearn.metrics import accuracy_score,classification_report

accuracy_score(y_pred,y_test)
```

```python
print(classification_report(y_pred,y_test))
```

### Pre-Prunning

#### Parameters

* define parameters with there range

```python
parameters ={
    'max_depth' : [1,2,3,4,5],
    'criterion': ['gini', 'entropy', 'log_loss'],
    'splitter' : ['best','random'],
    'max_features' : ['auto','sqrt','log2']
}
```

#### Cross validation check :

* to find the best parameter from the range provided

#### Create the treemodel

```python
treemodel_prp = DecisionTreeClassifier()
```

#### Create cross-validation (cv)

```python
from sklearn.model_selection import GridSearchCV

cv = GridSearchCV(treemodel_prp,param_grid=parameters,cv=5,scoring='accuracy')
```

#### Fit the cv

```python
cv.fit(X_train,y_train)
```

#### Find best params

```python
cv.best_params_
```

#### Now create tree with this params

```python
treemodel_prp_updated = DecisionTreeClassifier(max_depth=3,
                                               criterion='gini',
                                               max_features='sqrt',
                                               splitter='best')
```

#### Fit the treemodel_prp_updated

```python
treemodel_prp_updated.fit(X_train,y_train)
```

#### Predict the test data

```python
y_pred_prp = treemodel_prp_updated.predict(X_test)
```

#### Find the accuracy and classification report

```python
accuracy_score(y_test,y_pred_prp)
```

```python
print(classification_report(y_test,y_pred_prp))
```
